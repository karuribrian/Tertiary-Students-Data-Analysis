```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Read Excel file and extract data from column E
df = pd.read_excel(post)
column_b_counts = df.iloc[:, 1].value_counts()

# Generate random colors
num_bars = len(column_b_counts)
colors = plt.cm.viridis(np.linspace(0, 1, num_bars))

# Calculate percentages based on total count of all values
total_count = column_b_counts.sum()
column_b_percentages = (column_b_counts / total_count) * 100

# Plotting the column E distribution
plt.figure(figsize=(8, 5))  # Adjust figure size if needed
ax = column_b_counts.plot(kind='bar', color=colors)  # Use generated colors
plt.title('Distribution of Students by Age')
plt.xlabel('Age')
plt.ylabel('Counts')

# Annotate bars with counts and percentages
for i, (count, percentage) in enumerate(zip(column_b_counts, column_b_percentages)):
    plt.text(i, count + 0.1, f"\n{percentage:.1f}%", ha='center', va='bottom', rotation=90)
    plt.tight_layout()

plt.xticks(rotation=45)  # Rotate x-axis labels if needed
plt.show()
```

```python
import matplotlib.pyplot as plt
from collections import Counter

data = [
    "Leparua", "Nadung'oro", "Mutunyi", "Maili saba", "Ntalabany", "Leparua", "Leparua", "Timau", "Ntugi",
    "Murwa", "Subuiga", "Ntugi", "Subuiga", "Kiirua", "Ngarendare", "Ntugi", "Kilimani", "Mutunyi", "Rugusu",
    "Kilimani", "Leparua", "Bbc", "Kilimani", "Kanjoka", "Mutunyi", "Manyagalo", "Sang'a", "Subuiga",
    "Ntumburi", "Nanyuki", "Lmd", "Subuiga", "Ntugi", "Subuiga", "Ngarendare", "Ntugi", "Ntugi", "Ngaree sirkon",
    "Ntumburi", "Ntugi", "Ntumburi", "Leparua", "Maili saba", "Mutunyi"
]

# Count occurrences of each village/community
village_counts = Counter(data)

# Extract village/community names and their counts
villages = list(village_counts.keys())
counts = list(village_counts.values())

# Calculate total count
total_count = sum(counts)

# Calculate percentages
percentages = [(count / total_count) * 100 for count in counts]

# Plotting the bar graph with percentage labels
plt.figure(figsize=(10, 6))
bars = plt.bar(villages, counts, color=colors)
colors = ["red","skyblue","pink","green","yellow"]
# Add percentage labels on top of each bar
for bar, percentage in zip(bars, percentages):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             f'{percentage:.2f}%', ha='center', va='bottom', fontsize=8)

plt.xlabel('Village/Community')
plt.ylabel('Count')
plt.title('Distribution of Students Across Neighboring Communities of Lewa')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
quarterly_sighting_freq = [63, 2546, 2483, 25.46]
vehicle_deployments = [36000, 37817.20, 1817.20, 37817.20]
days_senior_officer = [200, 336, 136, 336]
foot_deployments = [5760, 8498.95, 2738.95, 8498.95]
aerial_patrols_hours = [24, 22.20, -1.8, 22.2]
rhinos_captured = [8, 10, 2, 10]
camera_traps_installed = [6, 6, 0, 6]

# Visualizations
plt.figure(figsize=(14, 18))

# Quarterly Sighting Frequency
plt.subplot(3, 2, 1)
plt.bar(quarters, quarterly_sighting_freq, color=['blue', 'green', 'red', 'purple'])
plt.title('Quarterly Sighting Frequency')
plt.xlabel('Quarter')
plt.ylabel('Sighting Frequency (%)')

# Vehicle Deployments
plt.subplot(3, 2, 2)
plt.bar(quarters, vehicle_deployments, color=['blue', 'green', 'red', 'purple'])
plt.title('Vehicle Deployments (km2)')
plt.xlabel('Quarter')
plt.ylabel('Distance Covered (km2)')

# Number of Days Senior Officer Takes on Ground
plt.subplot(3, 2, 3)
plt.bar(quarters, days_senior_officer, color=['blue', 'green', 'red', 'purple'])
plt.title('Number of Days Senior Officer on Ground')
plt.xlabel('Quarter')
plt.ylabel('Number of Days')

# Foot Deployments
plt.subplot(3, 2, 4)
plt.bar(quarters, foot_deployments, color=['blue', 'green', 'red', 'purple'])
plt.title('Foot Deployments (km)')
plt.xlabel('Quarter')
plt.ylabel('Distance Covered (km)')

# Aerial Patrols
plt.subplot(3, 2, 5)
plt.bar(quarters, aerial_patrols_hours, color=['blue', 'green', 'red', 'purple'])
plt.title('Aerial Patrols (hours)')
plt.xlabel('Quarter')
plt.ylabel('Duration (hours)')

# Number of Rhinos Captured
plt.subplot(3, 2, 6)
plt.bar(quarters, rhinos_captured, color=['blue', 'green', 'red', 'purple'])
plt.title('Number of Rhinos Captured')
plt.xlabel('Quarter')
plt.ylabel('Number of Rhinos')

plt.tight_layout()
plt.show()
```
