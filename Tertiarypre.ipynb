```python
import matplotlib.pyplot as plt
import pandas as pd

# File path
modified_file_path = "C:/Users/grants.support/Desktop/R TERTIARY ANALYSIS/PRE21.xlsx"

# Read Excel file
df = pd.read_excel(modified_file_path)

# Define function to plot donut chart
def plot_donut(column_index, title):
    # Extract data from column
    column_values = df.iloc[:, column_index].value_counts()
    
    # Generate colors
    colors = ['#1f77b4', '#ff7f0e']  # Blue for 'Yes' and orange for 'No'
    
    # Plot donut chart
    plt.figure(figsize=(4, 4))  # Adjust figure size if needed
    plt.pie(column_values, labels=column_values.index, autopct='%1.1f%%', startangle=90, colors=colors, wedgeprops=dict(width=0.4))
    plt.title(title)
    centre_circle = plt.Circle((0, 0), 0.3, color='white', linewidth=0)
    fig = plt.gcf()
    fig.gca().add_artist(centre_circle)
    plt.axis('equal')
    plt.show()

# Define function to plot pie chart
def plot_pie(column_index, title):
    # Extract data from column
    column_values = df.iloc[:, column_index].value_counts()
    
    # Plot pie chart
    plt.figure(figsize=(4, 4))  # Adjust figure size if needed
    plt.pie(column_values, labels=column_values.index, autopct='%1.1f%%', startangle=90)
    plt.title(title)
    plt.axis('equal')
    plt.show()

# Plotting donut and pie charts with specified titles
plot_donut(20, "Primary Expectations and Financial Planning in Mentorship Sessions")
plot_pie(21, "Primary Expectations and Networking Opportunities in Mentorship Sessions")
plot_donut(22, "Primary Expectations and Academic Support in Mentorship Sessions")
plot_pie(23, "Primary Expectations and Personal Development in Mentorship Sessions")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# File path
modified_file_path = "C:/Users/grants.support/Desktop/R TERTIARY ANALYSIS/PRE21.xlsx"

# Read Excel file
df = pd.read_excel(modified_file_path)

# Define the column indices and their corresponding labels
column_labels = {
    20: 'Financial Planning',
    21: 'Networking Opportunities',
    22: 'Academic Support',
    23: 'Personal Development'
}

# Define the column indices
relevant_column_indices = [20, 21, 22, 23]  # Assuming these are the indices for columns U, V, W, X

# Calculate percentage of '1's for each column
percentages = []
for column_index in relevant_column_indices:
    total_count = len(df.iloc[:, column_index])
    count_ones = df.iloc[:, column_index].sum()
    percentage = (count_ones / total_count) * 100
    percentages.append(percentage)

# Plotting the bar graph
plt.figure(figsize=(5, 4))
column_names = [column_labels[column_index] for column_index in relevant_column_indices]
bars = plt.bar(column_names, percentages, color=['blue', 'orange', 'green', 'red'])

# Adding data labels to the bars
for bar, percentage in zip(bars, percentages):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{percentage:.1f}%', 
             ha='center', va='bottom')

plt.title('Student Expectations Overview')
plt.xlabel('Expectations')
plt.ylabel('Percentage')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--')
plt.xticks(rotation=45)  # Rotate x-axis labels
plt.show()
```
